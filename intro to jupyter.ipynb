## How to write Python code in Jupyter?

As you'd expect really. You just tell Jupyter that you are entering code and then get going...

```python
# Exactly as you'd expect

a = 7

def fibonacci(n):
    if n < 1:
        return 1
    else:
        return n * fibonacci(n - 1)
```

- So we've defined a variable and a function
- Now we can use them: the scope travels between entry boxes

```python
print(f'The value of {fibonacci} for a = {a} is {fibonacci(a)}')
```

Now we can make some tweaks and then keep using the same variables and functions...

```python
while a > 0:
    print(f'The value of {fibonacci} for a = {a} is {fibonacci(a)}')
    a -= 1
```

You need to be a little careful though: if you run cells in the wrong order then you are in danger of creating infinite loops!

## Making things look nice

This gets really lovely when you realise that you can also include graphs using the excellent `matlotlib` library.

## Labelling your graphs

You can label and title your graphs, just as you were always taught to.

```python
from matplotlib import pyplot as plt

x = [5, 4, 3, 2, 1]
y = [1, 3, 5, 3, 1]

plt.clf()

plt.title('This is a scatter plot')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.scatter(x, y)
plt.show()

plt.title('This is a line plot')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.plot(x, y)
```

## This is how we can plot single sets of data, but we can do a lot more...

`matplotlib` is a beast of a package and, though it is pretty good for plotting simple sets of data, there is a lot of work to be done to get it behaving exactly as you would like.


```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

fig.clf()

ax = fig.add_subplot(111)

#scatter, = ax.scatter([1, 4, 5, 2], [2, 8, 3, 2])

x = [2, 1, 2, 4]
y = [5, 1, 1, 4]

line1, = ax.plot(x, y, c = 'r')

x2 = np.linspace(1, 6, 20)
y2 = np.sin(x2)

line2, = ax.plot(x2, y2, c = 'b')
```

```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 50)

y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
fig.clf()

ax = fig.add_subplot(111)

line1, = ax.plot(x, y1)
line2, = ax.plot(x, y2)

```
